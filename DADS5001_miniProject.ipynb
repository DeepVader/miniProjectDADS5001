# rice_yield_change

```python
df_filtered = df[(df["attribute"] == "ผลผลิตต่อเนื้อที่เก็บเกี่ยว") & (df["items"] == "จังหวัด")].copy()
df_filtered["values"] = pd.to_numeric(df_filtered["values"], errors="coerce")
df_filtered["year"] = pd.to_numeric(df_filtered["year"], errors="coerce")
display(df_filtered)
```

```python
df_yearly = df_filtered.groupby("year")["values"].sum().reset_index()
df_yearly["change (%)"] = df_yearly["values"].pct_change() * 100
display(df_yearly)
```

```python
colors = ['blue' if x >= 0 else 'red' for x in df_yearly["change (%)"]]

plt.figure(figsize=(12, 6))
bars = plt.bar(df_yearly["year"], df_yearly["change (%)"], color=colors, width=0.6, edgecolor='black')

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2,
             height + (0.05 if height >= 0 else -0.05),
             f"{height:.2f}%",
             ha="center",
             va="bottom" if height >= 0 else "top",
             fontsize=11,
             color="black",
             weight="bold")

plt.ylim(-5,5)
plt.axhline(0, color='black', linestyle='--', linewidth=1)
plt.title("อัตราการเปลี่ยนแปลงของผลผลิตข้าวรายปี", fontsize=16, fontweight='bold')
plt.xlabel("ปี", fontsize=13)
plt.ylabel("อัตราการเปลี่ยนแปลง (%)", fontsize=13)
plt.xticks(df_yearly["year"])
plt.grid(axis='y', linestyle='--', alpha=0.4)

import matplotlib.patches as mpatches
green_patch = mpatches.Patch(color='blue', label='เพิ่มขึ้นจากปีก่อน')
red_patch = mpatches.Patch(color='red', label='ลดลงจากปีก่อน')
plt.legend(handles=[green_patch, red_patch], loc='upper right', frameon=True, fontsize=11)

plt.tight_layout()
plt.savefig("images/rice_yield_change.png")
plt.show()
```

# rice_vs_rainfall_comparison

```python
df_filtered = df[(df["attribute"] == "ผลผลิต") & (df["items"] == "จังหวัด")]
df_filtered["year"] = pd.to_numeric(df_filtered["year"], errors='coerce')
df_filtered["ผลผลิตข้าว (ล้านตัน)"] = df_filtered["values"] / 1_000_000
display(df_filtered)
```

```python
rice_by_year = df_filtered.groupby("year")["ผลผลิตข้าว (ล้านตัน)"].sum().reset_index()
display(rice_by_year)
```

```python
rainfall_data = pd.DataFrame({
    "year": [2557, 2558, 2559, 2560, 2561, 2562, 2563, 2564, 2565],
    "ปริมาณน้ำฝน (มม.)": [1373, 1247, 1474, 1824, 1452, 1218, 1384, 1589, 1848]
})
display(rainfall_data)
```

```python
df_merged = pd.merge(rice_by_year, rainfall_data, on="year", how="inner")
display(df_merged)
```

```python
fig, ax1 = plt.subplots(figsize=(12, 6))

line1, = ax1.plot(df_merged["year"], df_merged["ผลผลิตข้าว (ล้านตัน)"],
                  marker="o", linestyle="-", color="tab:blue")
ax1.set_xlabel("ปี")
ax1.set_ylabel("ผลผลิตข้าว (ล้านตัน)", color="tab:blue")
ax1.tick_params(axis="y", labelcolor="tab:blue")
ax1.set_ylim(df_merged["ผลผลิตข้าว (ล้านตัน)"].min() - 1,
             df_merged["ผลผลิตข้าว (ล้านตัน)"].max() + 1)

for x, y in zip(df_merged["year"], df_merged["ผลผลิตข้าว (ล้านตัน)"]):
    ax1.text(x, y + 0.2, f"{y:.1f}", color="tab:blue", ha="center", fontsize=9)

ax2 = ax1.twinx()
line2, = ax2.plot(df_merged["year"], df_merged["ปริมาณน้ำฝน (มม.)"],
                  marker="s", linestyle="--", color="tab:green")
ax2.set_ylabel("ปริมาณน้ำฝน (มม.)", color="tab:green")
ax2.tick_params(axis="y", labelcolor="tab:green")
ax2.set_ylim(df_merged["ปริมาณน้ำฝน (มม.)"].min() - 50,
             df_merged["ปริมาณน้ำฝน (มม.)"].max() + 50)

for x, y in zip(df_merged["year"], df_merged["ปริมาณน้ำฝน (มม.)"]):
    ax2.text(x, y + 20, f"{y}", color="tab:green", ha="center", fontsize=9)

custom_lines = [
    Line2D([0], [0], color="tab:blue", marker="o", label="ผลผลิตข้าว (ล้านตัน)"),
    Line2D([0], [0], color="tab:green", marker="s", linestyle="--", label="ปริมาณน้ำฝน (มม.)")
]

plt.legend(handles=custom_lines,
           loc='lower right',
           fontsize=9,
           frameon=True,
           framealpha=1,
           facecolor='white',
           edgecolor='black',
           borderpad=1,)

ax1.grid(True, linestyle='--', linewidth=0.7, alpha=0.5)
plt.title("การเปรียบเทียบผลผลิตข้าวและปริมาณน้ำฝน", fontsize=13)
fig.tight_layout()
plt.savefig("images/rice_vs_rainfall_comparison.png")
plt.show()
```

# trend_rice(all)

```python
df_filtered = df[(df['area'] == 'ประเทศไทย') & (df['attribute'] == 'ผลผลิต')]
df_filtered = df_filtered[df_filtered['year'] <= 2565]
display(df_filtered)
```

```python
df_production = df_filtered.groupby('year')['values'].sum()
df_production = df_production / 1_000_000
display(df_production)
```

```python
fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.plot(df_production.index, df_production.values, marker='o', linestyle='-', color='b', label='ผลผลิต')
ax1.set_xlabel('ปี')
ax1.set_ylabel('ผลผลิต (ล้านตัน)', color='black')  # ปรับชื่อแกน Y เป็นล้านตัน
ax1.tick_params(axis='y', labelcolor='black')

ax1.grid(True, linestyle='--', linewidth=0.7, alpha=0.5)

for i in range(len(df_production)):
    ax1.annotate(f'{df_production.iloc[i]:,.2f}',
                 (df_production.index[i], df_production.iloc[i]),
                 textcoords="offset points", xytext=(0,-10), ha='center', fontsize=10, color='blue')

plt.legend(loc='upper right', fontsize=12, title='ผลผลิตข้าวรวม', title_fontsize=14,
           frameon=True, framealpha=1, facecolor='white', edgecolor='black', borderpad=1)
plt.title('แนวโน้มผลผลิตข้าวรวม (ข้าวนาปี + ข้าวนาปรัง) รายปี')
plt.grid(True)
plt.savefig("images/trend_rice(all).png")
plt.show()
```

# trend_rice_type

```python
df_filtered = df[(df['attribute'] == 'ผลผลิต') &
                 ((df['subcommod'] == 'ข้าวนาปี') | (df['subcommod'] == 'ข้าวนาปรัง')) &
                 (df['area'] == 'ประเทศไทย')]
display(df_filtered)
```

```python
df_production = df_filtered.groupby(['year', 'subcommod'])['values'].sum().unstack()
df_production = df_production / 1_000_000
display(df_production)
```

```python
fig, ax1 = plt.subplots(figsize=(10, 6))

df_production['ข้าวนาปี'].plot(ax=ax1, marker='o', linestyle='-', color='blue', label='ข้าวนาปี')

df_production['ข้าวนาปรัง'].plot(ax=ax1, marker='o', linestyle='-', color='orange', label='ข้าวนาปรัง')

ax1.set_xlabel('ปี', fontsize=12, fontweight='bold', color='black')
ax1.set_ylabel('ผลผลิต (ล้านตัน)', fontsize=12, fontweight='bold', color='black')
ax1.tick_params(axis='both', which='major', labelsize=10)
ax1.tick_params(axis='y', labelcolor='black')

ax1.set_xticks(df_production.index)

for i in range(len(df_production)):
    ax1.annotate(f'{df_production["ข้าวนาปี"].iloc[i]:,.2f}',
                 (df_production.index[i], df_production["ข้าวนาปี"].iloc[i]),
                 textcoords="offset points", xytext=(0, 5), ha='center', fontsize=10, color='blue')

    ax1.annotate(f'{df_production["ข้าวนาปรัง"].iloc[i]:,.2f}',
                 (df_production.index[i], df_production["ข้าวนาปรัง"].iloc[i]),
                 textcoords="offset points", xytext=(0, 5), ha='center', fontsize=10, color='orange')

ax1.grid(True, linestyle='--', linewidth=0.7, alpha=0.5)

plt.title('แนวโน้มผลผลิตข้าวรายปี', fontsize=14, fontweight='bold')

plt.legend(title='ชนิดของข้าว', loc='upper left', fontsize=12, title_fontsize=14,
           frameon=True, framealpha=1, facecolor='white', edgecolor='black', borderpad=1, bbox_to_anchor=(0.8, 0.9))

plt.tight_layout()
plt.savefig("images/trend_rice_type.png")
plt.show()
```

# trend_rice_region

```python
df_filtered = df[(df['attribute'] == 'ผลผลิต') &
                 (df['area'].isin(['ภาคเหนือ', 'ภาคใต้', 'ภาคตะวันออกเฉียงเหนือ', 'ภาคกลาง']))]
display(df_filtered)
```

```python
df_production = df_filtered.groupby(['year', 'area'])['values'].sum().unstack()
df_production = df_production / 1_000_000
display(df_production)
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

colors = ['green', 'blue', 'red', 'orange']
areas = ['ภาคตะวันออกเฉียงเหนือ', 'ภาคเหนือ', 'ภาคกลาง', 'ภาคใต้']

sorted_areas = df_production.max(axis=0).sort_values(ascending=False).index.tolist()

for i, area in enumerate(sorted_areas):
    ax.plot(df_production.index, df_production[area], marker='o', linestyle='-', color=colors[i], label=area)

    for j in range(len(df_production)):
        ax.annotate(f'{df_production[area].iloc[j]:,.2f}',
                    (df_production.index[j], df_production[area].iloc[j]),
                    textcoords="offset points", xytext=(0, 5), ha='center', fontsize=10, color=colors[i])

ax.set_xlabel('ปี', fontsize=12, fontweight='bold', color='black')
ax.set_ylabel('ผลผลิต (ล้านตัน)', fontsize=12, fontweight='bold', color='black')
ax.tick_params(axis='both', which='major', labelsize=10)
ax.tick_params(axis='y', labelcolor='black')

ax.set_xticks(df_production.index)

ax.grid(True, linestyle='--', linewidth=0.7, alpha=0.5)

plt.title('เปรียบเทียบผลผลิตข้าวในพื้นที่ในแต่ละภูมิภาค (ภาคตะวันออกเฉียงเหนือ, ภาคเหนือ, ภาคกลาง, ภาคใต้) รายปี ', fontsize=14, fontweight='bold')

plt.legend(title='Area', loc='center left', fontsize=12, title_fontsize=14,
           frameon=True, framealpha=1, facecolor='white', edgecolor='black', borderpad=1, bbox_to_anchor=(0.76, 0.28))


plt.tight_layout()
plt.savefig("images/trend_rice_region.png")
plt.show()
```

# top10_province_2565

```python
df_filtered = df[(df['attribute'] == 'ผลผลิต') & (df['year'] == 2565) & (df['items'] == 'จังหวัด')]
display(df_filtered)
```

```python
df_production_by_province = df_filtered.groupby(['area'])['values'].sum().reset_index()
display(df_production_by_province)
```

```python
df_top_10_provinces = df_production_by_province.groupby('area')['values'].sum().nlargest(10)
df_top_10_provinces = df_top_10_provinces / 1_000_000
display(df_top_10_provinces)
```

```python
plt.figure(figsize=(12, 6))

colors = plt.cm.Greens(range(0, 256, int(256/10)))
ax = df_top_10_provinces.sort_values(ascending=True).plot(kind='barh', color=colors, edgecolor='black')

for rect in ax.patches:
    width = rect.get_width()
    ax.text(width -0.5, rect.get_y() + rect.get_height()/2, f'{width:,.2f}',
            va='center', ha='right', fontsize=12, color='black')


plt.title('Top 10 อันดับที่ผลิตข้าวได้มากที่สุดในปี 2565', fontsize=14, fontweight='bold')
plt.xlabel('ผลผลิต (ล้านตัน)', fontsize=12, fontweight='bold')
plt.ylabel('จังหวัด', fontsize=12, fontweight='bold')
plt.tight_layout()
plt.savefig("images/top10_province_2565.png")
plt.show()
```

# rice_per_caltivated_region

```python
df_filtered = df.query("attribute == 'ผลผลิตต่อเนื้อที่เพาะปลูก' & items == 'ภาค' ")[["year", "area", "attribute", "values", "subcommod"]]
display(df_filtered)
```

```python
df_production_by_province = df_filtered.groupby(["year", "area", "attribute"])["values"].sum().reset_index()
df_production_by_province = df_production_by_province.groupby("area")["values"].mean().reset_index()
df_production_by_province = df_production_by_province.sort_values(by="values", ascending=True)
display(df_production_by_province)
```

```python
plt.figure(figsize=(12, 6))

colors = plt.cm.Greens(range(0, 256, int(256/10)))

ax = df_production_by_province.plot(kind='barh', color="orange", edgecolor='black')

for rect in ax.patches:
    width = rect.get_width()
    ax.text(width - 0.05, rect.get_y() + rect.get_height()/2, f'{width:,.0f}',
            va='center', ha='right', fontsize=12, color='black')

plt.title('ผลผลิตต่อเนื้อที่เพาะปลูกเฉลี่ยตามภูมิภาคปี 2557-2565', fontsize=14, fontweight='bold')
plt.xlabel('ผลผลิตต่อเนื้อที่เพาะปลูกเฉลี่ย (ตัน)', fontsize=12, fontweight='bold')
plt.ylabel('ภาค', fontsize=12, fontweight='bold')

plt.legend().set_visible(False)

ax.set_yticklabels(df_production_by_province['area'], fontsize=12, fontweight='bold')

plt.tight_layout()
plt.savefig("images/rice_per_caltivated_region.png")
plt.show()
```
